# ! Welcome to Matplotlib tutorial !

###  1) Matplotlib is a cross-platform, data visualization and graphical plotting library for Python and its numerical extension NumPy. As such, it offers a viable open source alternative to MATLAB. Developers can also use matplotlib's APIs (Application Programming Interfaces) to embed plots in GUI applications.
###  2) Matplotlib is a low level graph plotting library in python that serves as a visualization utility.

```python
#import
import matplotlib.pyplot as plt
```

###  Line Plot

```python
x=[1,2,3,4]
y=[5,4,3,1]
plt.plot(x,y)
```

```python
#with marker=o
x=[1,2,3,4]
y=[4,3,5,1]
plt.plot(x,y,marker="o")
```

```python
#with marker=*
x=[1,2,3,4]
y=[4,3,5,1]
plt.plot(x,y,marker="*")
```

```python
#with marker=+
x=[1,2,3,4]
y=[4,3,5,1]
plt.plot(x,y,marker="+")
```

```python
#with color=r,g,y,b,v
x=[2,3,4,5]
y=[3,1,4,5]
plt.plot(x,y,color='r',marker='o')
```

```python
#with label
d=[2,4,5,6]
e=[1,3,4,5]
plt.plot(d,e,marker="*",color='b',label="myline")
plt.legend()
#legend must to get a label
```

```python
a=[1,3,5,6]
b=[2,3,5,6]
plt.plot(a,b,marker="*",color="g",label="x-y")
plt.legend()
plt.plot(b,a,marker="o",color="k",label="b-a")
plt.legend()
#BY REVERSING A & B WE CAN PLOT TWO DIFFERENT LINES IN SAME PLOT
```

```python
#WITH TITLE
G=[1,6,3,4]
H=[2,3,4,6]
plt.plot(G,H,marker="o",label="xaxis",color="g")
plt.title("MY LINE PLOT")
plt.legend()
```

```python
#x label and y label
G=[1,6.2,3.3,7]
H=[0,3.6,5,8]
plt.plot(G,H,marker="o",label="xaxis",color="g")
plt.xlabel("G")
plt.ylabel("H")
plt.title("MY LINE PLOT-2")
plt.legend()
```

```python
#with new line style
G=[1,4,4.5,5]
H=[2.3,3.5,4.6,6.5]
plt.plot(G,H,marker="o",label="xaxis",color="g",linestyle="--")
plt.title("MY LINE PLOT")
plt.legend()
```

```python
#marker face color
G=[1,5,3,7]
H=[2,3,4,6]
plt.plot(G,H,marker="o",label="xaxis",color="y",linestyle="--",markerfacecolor="b")
plt.title("PLOT")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.legend()
```

```python
#assigning values to x and y 
plt.plot(x,y,marker="o",color="b",linestyle="--",markerfacecolor="k",label="tickschange")
plt.legend()
plt.xticks([1,2,3,4])
plt.yticks([1.2,2.4,3.6,4.2])
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.legend()
```

```python
#changing position of a label
plt.plot(x,y,marker="o",color="k",linestyle="--",markerfacecolor="b",label="loc-change")
plt.legend(loc="best")
plt.xticks([1.1,2.1,3.1,4.1])
plt.yticks([1.4,2.4,3.4,4.4])
plt.xlabel("x-axis")
plt.ylabel("y-axis")
```

```python
#fontsize and upper left
plt.plot(x,y,marker="o",color="g",markerfacecolor="k",label="Upper-left")
plt.legend(loc="upper left")
plt.xlabel("X-ax")
plt.ylabel("Y-ax")
plt.title("LS",fontsize=20)
```

```python
#fontsize and upper right
plt.plot(x,y,marker="o",color="g",markerfacecolor="k",label="Upper-right")
plt.legend(loc="upper right")
plt.xlabel("X-ax")
plt.ylabel("Y-ax")
plt.title("LS",fontsize=18.5)
```

```python
#fontsize and upper centre
plt.plot(x,y,marker="o",color="g",markerfacecolor="k",label="up-center")
plt.legend(loc="upper center")
plt.xlabel("X-ax")
plt.ylabel("Y-ax")
plt.title("LS",fontsize=17.4)
```

```python
#fontsize and lower left
plt.plot(x,y,marker="o",color="g",markerfacecolor="k",label="low left")
plt.legend(loc="lower left")
plt.xlabel("X-ax")
plt.ylabel("Y-ax")
plt.title("LS",fontsize=16.4)
#as above we can use lower left,lower right,lower center
```

```python
a=[1,4,3,7,2,6,5]
b=["mon","tue","thu","wed","fri","sat","sun"]
#plt.plot(a,b)
plt.plot(a,b,marker="*",color="k",label="wk day graph",linestyle="--",markerfacecolor="b")
plt.legend(loc="lower center")
plt.title("wk day graph")
plt.xlabel("x axis")
plt.ylabel("y axis")
```

## use in datasets

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# use Teams data
# bar plot of all teams points total
# x lable -> team name
 # y-> total
# line plot of all team ranks points with lables(legend)
#x->year
#y->points

#scatter plot of rank and points rank vs points
df=pd.read_csv("C:\\Users\\Pratik1\Desktop\dataset\\data's\\Teams.csv")
df
```

```python
teams=df.groupby('Team')['Points'].sum()
plt.bar(range(4),teams)
plt.xticks(range(4),teams.index)
plt.xlabel("Teams")
plt.ylabel("Total")
#plt.bar(teams,marker="o",color="g",markerfacecolor="k")
```

```python
teams=df.groupby("Team")["Rank"].sum()
plt.plot(range(4),teams,color="g",label="x-ax",marker="o",markerfacecolor="k")
plt.xticks(range(4),teams.index)
plt.xlabel("Year")
plt.ylabel("Points")
plt.legend()
```

## GRID- Creates a bagroud to plot

```python
teams=df.groupby("Team")["Rank"].sum()
plt.plot(range(4),teams,color="g",label="x-ax",marker="o",markerfacecolor="k")
plt.xticks(range(4),teams.index)
plt.xlabel("Year")
plt.ylabel("Points")
plt.grid(color = 'green', linestyle = '--', linewidth = 0.5)
plt.legend()
#grid can customized by use of x or y , both
```

## COLORBAR

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
colors = np.array([0, 10, 20, 30, 40, 45, 50, 55, 60, 70, 80, 90, 100])

plt.scatter(x, y, c=colors, cmap='rainbow_r')

plt.colorbar()

plt.show()
```

## PIECHART

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])

plt.pie(y)
plt.show() 
```

```python
#LABELS
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels = mylabels)
plt.show() 
```

```python
#STARTANGLE IS DEFAULT-X AXIS BUT WE CAN CHANGE IT
plt.pie(y, labels = mylabels, startangle = 90)
plt.show() 
```

```python
#EXPLODE=can be cut out some part
myexplode = [0.2, 0, 0, 0]

plt.pie(y, labels = mylabels, explode = myexplode)
plt.show() 
```

```python
#shadow =true/false
plt.pie(y, labels = mylabels, explode = myexplode, shadow = True)
plt.show() 
```

```python
#changing a colors
mycolors = ["black", "hotpink", "b", "#4CAF50"]
plt.pie(y, labels = mylabels, colors = mycolors,shadow=True,explode=myexplode)
plt.show() 
```

```python
#legend=To add a list of explanation for each wedge
plt.pie(y, labels = mylabels,colors=mycolors)
plt.legend()
plt.show() 
```

```python
#legend with header
plt.pie(y, labels = mylabels)
plt.legend(title = "Four Fruits:")
plt.show() 
```
